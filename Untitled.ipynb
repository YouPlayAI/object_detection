```python
criterion = losses.create_losses(config.neg_ratio, NUM_CLASSES)
```

```python
for filename, imgs, gt_confs, gt_locs in batch_generator.take(1):
    print("filename", filename)
    print("image shape", imgs.shape)
    print("class label", gt_confs)
    print("location label", gt_locs)
    
    confs, locs = ssd(imgs)
    
    print("predicted confs :", confs)
    print("predicted locations :", locs)

    conf_loss, loc_loss = criterion(tf.argmax(gt_confs, axis=-1), gt_locs, gt_confs, gt_locs)
    print("conf loss :", conf_loss)
    print("loc loss :", loc_loss)
    print("-------------------------------------------------")
```
